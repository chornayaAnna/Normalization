```python
# удаление столбца 
form.drop(['Отметка времени'], axis = 1, inplace = True)
```

```python
form.head(5)
```

```python
# переимнование столбца
form = form.rename(columns = {'Сколько вам лет?' : 'age'})
```

```python
form.head(5)
```

Бинаризация

```python
form['children'] = (form['Есть ли у вас дети?'] == "да").astype(int)
```

```python
form.head(11)
```

```python
form['price'] = form['Вас устраивают цены на товары?'].apply(lambda x: 0 if x == "нет" else 1)
```

```python
form
```

```python
form['car'] = np.where(form['Есть ли у вас автомобиль?'] == 'да', 1, 0)
```

```python
form
```

Категориальные признаки

```python
form['locality'] = form['В какой местности вы живете?'].map({'город' : 1, 'село' : 2, 'поселок' : 3, 'пригород' : 4})
```

```python
form
```

```python
# заполнение пустых значений
#form['locality'] = form['locality'].fillna(0)
form['locality'] = round(form['locality'].fillna(form['locality'].mean()))
form.head(36)
```

```python
# преобразуем тип
form['locality'] = form['locality'].astype('int')
```

```python
form
```

```python
# удаление пустых значений
form.dropna()
```

```python
form['female'] = (form['Ваш пол:'] == "женский").astype(int)
```

```python
form['male'] = (form['Ваш пол:'] == "мужской").astype(int)
```
